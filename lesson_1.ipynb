```python
xs = np.array([1, 2, 7, 4, 5, 6, 7], dtype=np.float32)
ys = np.array([7, 6, 5, 2, 3, 2, 1], dtype=np.float32)
```

```python
# Вариант 1
N = xs.size
s = []
for x, y in zip(xs, ys):
    q = (y-x)**2
    s.append(q)

result = np.sum(s)/N

print(result)
```

```python
# Вариант 2
N = xs.size
s = (ys - xs)**2
result = np.sum(s)/N

print(result)
```

```python
# Вариант 3
N = xs.size
s = ys - xs
result = (s @ s)/N

print(result)
```

```python
xs = np.array([1, 2, 7, 4, 5, 6, 7], dtype=np.float32)
ys = np.array([7, 6, 5, 2, 3, 2, 1], dtype=np.float32)

N = xs.size
result = (ys @ xs)/N

print(result)
```

```python
xs = np.array([5, 2, 7, 4, 5, 6, 7], dtype=np.float32)
errors = np.array([7, 6, 5, 2, 3, 2, 1], dtype=np.float32)
```

```python
# Вариант 1
r_1 = -2.0 * np.mean(xs * errors)
r_2 = -2.0 * np.mean(errors)
result = np.array([r_1, r_2])

print(result)
```

```python
# Вариант 2
new_xs = np.stack((xs, np.ones_like(xs)))

print(new_xs)
print(new_xs.shape)
```

```python
N = xs.shape[0]
result = -2.0 * (errors @ new_xs / N)   # (2,)

print(result)
```
