```python
import math
import numpy as np 
np.__version__
```

```python
def sigmoid(x: float) -> float:
    return 1/(1 + math.exp(-x))


def func(x: float) -> float:
    return x**3 + 5*x**2 - math.sin(x**2)
```

```python
x = 0.1
sgmd = sigmoid(x)

# Аналитическое значение производной.
d_sgmd_1 = sgmd * (1-sgmd)
print("Аналитическое значение производной:")
print(f"{d_sgmd_1:.10f}")

# Значение производной по определению.
d_xs = np.geomspace(0.1, 0.000001, 6)
print("\nЗначение производной по определению:")
for dx in d_xs:
    d_sgmd_2 = (sigmoid(x+dx) - sigmoid(x))/dx
    print(f"{d_sgmd_2:.10f} при dx={dx:.6f}")
    
```

```python
x_0 = 0.0
sgmd_0 = sigmoid(x_0)

# Значение производной в точке x_0.
d_sgmd = sgmd_0 * (1-sgmd_0)
sign = "+" if d_sgmd > 0 else "-"

# Корректируем x_0 чтобы уменьшить значение sigmoid.
lr = 0.0001
x_1 = x_0 - lr * (1.0 if d_sgmd > 0.0 else -1.0)

sgmd_1 = sigmoid(x_1)

print(f"sgmd_0={sgmd_0:.10f}, x_0={x_0:.4f}")
print(f"Знак производной '{sign}', d_sgmd={d_sgmd:.5f}")
print(f"sgmd_1={sgmd_1:.10f}, x_1={x_1:.4f}")

```

```python
x = 0.1
f = func(x)

# Аналитическое значение производной.


# Значение производной по определению.
d_xs = np.geomspace(0.1, 0.000001, 6)
print("Значение производной по определению:")
for dx in d_xs:
    d_f = (func(x+dx) - func(x))/dx
    print(f"{d_f:.10f} при dx={dx:.6f}")
    
```

```python
x_0 = -3.0
f_0 = func(x_0)

# Значение производной в точке x_0.
dx = 0.00001
d_f = (func(x_0+dx) - func(x_0))/dx
sign = "+" if d_f > 0 else "-"

# Корректируем x_0, чтобы уменьшить значение func.
lr = 0.0001
x_1 = x_0 - lr * (1.0 if d_f > 0.0 else -1.0)

f_1 = func(x_1)

print(f"f_0={f_0:.5f}, x_0={x_0:.4f}")
print(f"Знак производной '{sign}'")
print(f"f_1={f_1:.5f}, x_1={x_1:.4f}")

```
